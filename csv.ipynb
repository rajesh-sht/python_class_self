```python
# DataFrame
b = {}
rol = []
na = []
ag = []
add = []
n = int(input("Enter how many rows = "))
for i in range(1,n+1):
    name = input("Enter name = ")
    age = int(input("Enter age = "))
    address = input("Enter address = ")
    rol.append(i)
    na.append(name)
    ag.append(age)
    add.append(address)

b = {'Roll':rol,'Name':na,'Age':ag,'Address':add}
import pandas as pd
data_frame = pd.DataFrame(b)
data_frame.to_csv('info.csv')
```
