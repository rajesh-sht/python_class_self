```python
# ------------------------------------Assignment(Nepal stock-Floorsheet)-------------------------------
from bs4 import BeautifulSoup
import requests
import pandas as pd

n = int(input("Enter page no = "))
final_data = []
for i in range(1,n+1):
    url = BeautifulSoup(f'http://www.nepalstock.com/main/floorsheet/index/{i}/', 'html.parser')
    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data,'xml')
    tags = soup.table
    tag = tags.find_all('tr')
    if (i==1): #for page first, include header
        tag = tag[1:-3]
    else: #for page other than first, exclude header
        tag = tag[2:-3]
        
    list_data = []
    for i in range(len(tag)):
        list_data.append(tag[i].get_text())
  
    for i in list_data:
        final_data.append(i.split('\n')[1:-1])

    
# write final_data into csv file
my_dataframe = pd.DataFrame(final_data)
my_dataframe.to_csv('floorchart.csv',mode = 'w',header = False,index = None)

# to read csv file
df = pd.read_csv('floorchart.csv',index_col = 'S.N.')
df
```

```python
# ------------------------------------Assignment(Nepal stock-Today's Price)-------------------------------
from bs4 import BeautifulSoup
import requests
import pandas as pd

n = int(input("Enter page no = "))
final_data = []
for i in range(1,n+1):
    url = BeautifulSoup(f'http://www.nepalstock.com/main/todays_price/index/{i}/', 'html.parser')
    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data,'xml')
    tags = soup.table
    tag = tags.find_all('tr')
    if (i==1): #for page first, include header
        tag = tag[1:-3]
    else: #for page other than first, exclude header
        tag = tag[2:-3]
        
    list_data = []
    for i in range(len(tag)):
        list_data.append(tag[i].get_text())
  
    for i in list_data:
        final_data.append(i.split('\n')[1:-1])

    
# write final_data into csv file
my_dataframe = pd.DataFrame(final_data)
my_dataframe.to_csv('floorchart.csv',mode = 'w',header = False,index = None)

# to read csv file
df = pd.read_csv('floorchart.csv',index_col = 'S.N.')
df
```
