```python
from bs4 import BeautifulSoup
import requests

soup = BeautifulSoup("https://www.worldometers.info/coronavirus/", "html.parser")
response = requests.get(soup)
data = response.text

datas = BeautifulSoup(data,'lxml')
tags = datas.table
tag = tags.find_all('tr')

# creating empty list
list_data = []
for i in range(len(tag)):
    list_data.append(tag[i].get_text())

final_data = []
for i in list_data:
    final_data.append(i.split('\n')[1:-3])

# delete unwanted datas from list
unwanted_index = [1,2,3,4,5,6,7,8,230,231,232,233,234,235,236]
for element in sorted(unwanted_index, reverse = True):
    del final_data[element]

# merging two columns in one
final_data[0][13] = final_data[0][13]+final_data[0][14]
# deleting unnecessary elements from list
del final_data[0][14:16]
del final_data[-1][-4:-1]

# data re-arrange in list
final_data.insert(83,final_data[221])
del final_data[-2]
final_data
for i in final_data[83:-1]:
    if i[0] == '221':
        i[0] = '83'
    else:
        i[0] = str(int(i[0])+1)

import pandas as pd
# # write final_data into csv file
my_dataframe = pd.DataFrame(final_data)
my_dataframe.to_csv('worldometers.csv',mode = 'w',header = False,index = None)

# to read csv file
df = pd.read_csv('worldometers.csv',index_col = '#')
df
```
